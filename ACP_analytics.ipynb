```python
import seaborn as sns
```

```python
PRE_proc = PRE.dropna()
ARE_proc = ARE.dropna()
SUB_proc = SUB.dropna()
STAN_proc = STAN.dropna()
LND_proc = LND.dropna()
LCAT_proc = LCAT.dropna()
CAT_proc = CAT.dropna()
LMENU_proc = LMENU.dropna()
MENU_proc = MENU.dropna()
PUR_proc = PUR.dropna()
```

```python
PRE_proc['type'] = "PRE"
ARE_proc['type'] = "ARE"
SUB_proc['type'] = "SUB"
STAN_proc['type'] = "STAN"
LND_proc['type'] = "LND"
LCAT_proc['type'] = "LCAT"
CAT_proc['type'] = "CAT"
LMENU_proc['type'] = "LMENU"
MENU_proc['type'] = "MENU"
PUR_proc['type'] = "PUR"
```

```python
data = pd.concat([PRE_proc, ARE_proc])
data = pd.concat([data, SUB_proc])
data = pd.concat([data, STAN_proc])
data = pd.concat([data, LND_proc])
data = pd.concat([data, LCAT_proc])
data = pd.concat([data, CAT_proc])
data = pd.concat([data, LMENU_proc])
data = pd.concat([data, MENU_proc])
data = pd.concat([data, PUR_proc])
```

```python
sns.pairplot(data, hue ="type" )
```

```python
sns.stripplot(x='type', y='reserve_rate', data=data)
```

①地図を絞れば絞るほど、予約率が向上していく。
タイプ毎に相対的な予約率が見える。

```python
sns.lmplot(x='PV_per_ACP', y='n_reserve', hue='type', data=data,fit_reg=False)
```
